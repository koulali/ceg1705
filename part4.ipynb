$$
\phi = tan^{-1}\left( \dfrac{Z + \varepsilon b \sin^3u}{p-e^2 a \cos^3{u}} \right)
$$

where:

$$
p = \sqrt{X^2+Y^2}
$$

$$
u = tan^{-1}\left( \frac{Za}{pb}\right)
$$

$$
\varepsilon = \frac{e^2}{1-e^2}
$$
